```python
##scrap designation

designation = browser.find_elements(By.CLASS_NAME,value='job-card-list__title')
a=[]
for i in designation:
    a.append(i.text)
print(a)
print()
print(len(a))
print()
print()

#scrap location

location = browser.find_elements(By.CLASS_NAME,value="job-card-container__metadata-item")
b=[]
for i in location:
    b.append(i.text)
    
print(b)
print()
print(len(b))
print()

c=[]
for i in b:
    if i!='On-site' and i!='Remote' and i!='Hybrid':
        c.append(i)
    else:
        pass
print(c)
print()
print(len(c))
print()
print()

# scrap companyname 

company = browser.find_elements(By.CLASS_NAME,value="job-card-container__company-name")
d=[]
for i in company:
    d.append(i.text)
    
print(d)
print()
print(len(d))

```

```python
#scrapping company link for industryname,no_of followers,no_emp

rt = browser.find_elements(By.CLASS_NAME ,value="job-card-container__company-name")
w=[]
for i in rt:
    w.append(i.get_attribute("href"))
print(w)
print()
print(len(w))
```

```python
#concat the link for no_emp

eer = []
for u in w:
    pqr = str(u) + 'people/'
    eer.append(pqr)
print(eer)
print()
print(len(eer))
```

```python
##scrapping link for level,no_applicant

poo = browser.find_elements(By.CLASS_NAME ,value="job-card-list__title")
la=[]
for lk in poo:
    la.append(lk.get_attribute("href"))
print(la)
print()
print(len(la))
```

```python
#scrapping industryname

import time 
qw=[]
for o in w:
    try:
        browser.get(o) 
        time.sleep(5)
        lo = browser.find_elements(By.XPATH,'//*[@id="ember26"]/div[2]/div[1]/div[1]/div[2]/div/div/div[1]')
        for p in lo:
            qw.append(p.text)
    except:
        qw.append('None')
print(qw)
print()
print(len(qw))
```

```python
#scrapping no_followers

import time
No_followers=[]
for y in w:
    try:
        browser.get(y)
        time.sleep(5)
        op = browser.find_elements(By.XPATH,'//*[@id="ember26"]/div[2]/div[1]/div[1]/div[2]/div/div')
        for m in op:
            No_followers.append(m.text)
    except:
        No_followers.append('None')
        
print(No_followers)
print()
print(len(No_followers))
```

```python
#scrapping the no_emp

import time 
no_emp=[]
for y in eer:
    try:
        browser.get(y)
        time.sleep(5)
        op = browser.find_elements(By.XPATH,'//*[@id="main"]/div[2]/div/div[1]/div[1]/h2')
        for m in op:
            no_emp.append(m.text)
    except:
        no_emp.append('None')
        
print(no_emp)
print()
print(len(no_emp))
```

```python
# scrapping no_applicant

import time 
no_applicant=[]
for o in la:
    try:
        browser.get(o)
        time.sleep(5)
        opo = browser.find_elements(By.XPATH , '/html/body/div[5]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[1]/span[2]/span[2]')
        for kl in opo:
            no_applicant.append(kl.text)
    except:
        no_applicant.append('None')
        
print(no_applicant)
print()
print(len(no_applicant))
```

```python
# scrapping level

import time 
level=[]
for o in la:
    try:
        browser.get(o)
        time.sleep(5)
        opoo = browser.find_elements(By.XPATH ,'/html/body/div[5]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[2]/ul/li[1]/span')
        for kll in opoo:
            level.append(kll.text)
    except:
        level.append('None')
        
print(level)
print()
print(len(level))
```
